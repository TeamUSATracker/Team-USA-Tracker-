```python
#Python Web Scrapping
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
num_list = [42, 43, 44,45, 46, 47,49, 58,60,62]
```

```python
winter_olympics = []
winter_olympics_all = pd.DataFrame()
for i in num_list:
    url = 'https://www.olympedia.org/editions/' + str(i) + '/medal'
    html = urlopen(url)
    bs = BeautifulSoup(html.read(), 'lxml')
    tables = bs.find_all('table',{"class":"table"})
    results1 = pd.read_html(str(tables))
    df = results1[0]
    df['Winter Olympics'] = i
   
    df.loc[df.iloc[:, 2].str.len() > 3, 'Sport'] = df.iloc[:, 2] #if Unnamed:2 contains a value with more than 3 characters, then this is the sport - set the value in a new Sport column
    df['Sport'] = df['Sport'].fillna(method='ffill') #fill in NA values using the previous non-NA value
    winter_olympics_all = pd.concat([winter_olympics_all,df], ignore_index=True) #concat with previous dataframe
    winter_olympics.append(df)
```

```python

#ReName Country Gold Column
winter_olympics_all.rename(columns={'Unnamed: 2':'Country Gold'},inplace=True)
winter_olympics_all.rename(columns={'Unnamed: 4':'Country Silver'},inplace=True)
winter_olympics_all.rename(columns={'Unnamed: 6':'Country Bronze'},inplace=True)

#Concat Sport/Event and Sport 
winter_olympics_all["Event"] = winter_olympics_all["Sport"] + ': ' + winter_olympics_all["Sport/Event"]

winter_olympics_all=winter_olympics_all[winter_olympics_all["Sport/Event"]!=winter_olympics_all["Gold"]]
winter_olympics_all
```
